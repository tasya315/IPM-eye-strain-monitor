```python
import cv2 as cv
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, Dense, BatchNormalization, Dropout, Flatten
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
```

```python
label_types = ['open', 'blink']
imdirs = ['./eyeDataset/open', './eyeDataset/closed']
images = []
labels = []

for i, imdir in enumerate(imdirs):
    for filename in os.listdir(imdir):
        img = cv.imread(os.path.join(imdir, filename))
        if img is not None:
            images.append(img)
            labels.append(i)

images = np.array(images)
# raw_dataset = np.array([images, labels], dtype="object")
# np.random.shuffle(raw_dataset)
# print(images[:,1])
```

```python
# images = np.reshape(images, (images))
print(images.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.3, random_state=42)
X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=0.66, random_state=42)

y_train = np.array(y_train)
y_valid = np.array(y_valid)

print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

```python
input_shape = keras.Input(shape=(24, 24, 3))
res_model = ResNet50(include_top=False, weights="imagenet", input_tensor=input_shape)
```

```python
res_model.summary()
```

```python
for layer in res_model.layers[:143]:
    layer.trainable=False
```

```python
for i, layer in enumerate(res_model.layers):
    print(i, layer.name, "-", layer.trainable)
```

```python
model = Sequential()
model.add(res_model)
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(32, activation="relu"))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(2, activation="sigmoid"))
model.compile(loss="binary_crossentropy",
             optimizer=RMSprop(learning_rate=0.00002),
             metrics=["accuracy"])
model.summary()
```

```python
checkpoint = ModelCheckpoint(filepath="test.h5", monitor="val_acc", mode="max", save_best_only=True)
```
