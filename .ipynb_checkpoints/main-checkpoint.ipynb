```python
import cv2
import math
import numpy as np
import mediapipe as mp
from rotated_rect_crop import crop_rotated_rectangle

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()

input_dim = (23,23)

cap = cv2.VideoCapture(1)
with mp_face_mesh.FaceMesh(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as face_mesh:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      continue

    height, width, _ = image.shape
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    result = face_mesh.process(rgb_image)
    if result.multi_face_landmarks:
      for facial_landmarks in result.multi_face_landmarks:
          pt1 = facial_landmarks.landmark[130]
          xi1 = int(pt1.x * width)
          yi1 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[133]
          xi2 = int(pt1.x * width)
          yi2 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[145]
          xi3 = int(pt1.x * width)
          yi3 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[159]
          xi4 = int(pt1.x * width)
          yi4 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[263]
          xa1 = int(pt1.x * width)
          ya1 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[362]
          xa2 = int(pt1.x * width)
          ya2 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[374]
          xa3 = int(pt1.x * width)
          ya3 = int(pt1.y * height)
          pt1 = facial_landmarks.landmark[386]
          xa4 = int(pt1.x * width)
          ya4 = int(pt1.y * height)
          
          m = (yi1-ya1)/(xi1-xa1)
          angle_in_radians = math.atan(m)
          angle_in_degrees = math.degrees(angle_in_radians)
    
          width = max(abs(xi1-xi2)+10, abs(yi3-yi4)+10)
          rect = (((xi1+xi2)//2, (yi1+yi2)//2), (width, width), angle_in_degrees)
          box = cv2.boxPoints(rect)
          box = np.int0(box)
          cv2.drawContours(image,[box],0,(255,255,255),1)
          image_cropped = cv2.resize(crop_rotated_rectangle(image, rect), input_dim)
          gray_image = cv2.cvtColor(image_cropped, cv2.COLOR_BGR2GRAY)
          gray_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB)
#           cv2.imshow("right", gray_image)

        
          width = max(abs(xa1-xa2)+10, abs(ya3-ya4)+10)
          rect = (((xa1+xa2)//2, (ya1+ya2)//2), (width, width), angle_in_degrees)
          box = cv2.boxPoints(rect)
          box = np.int0(box)
          cv2.drawContours(image,[box],0,(255,255,255),1)
          image_cropped = cv2.resize(crop_rotated_rectangle(image, rect), input_dim)
#           cv2.imshow("left", image_cropped)

    cv2.imshow("Image", image)
    if cv2.waitKey(1) &  0xFF == ord('q'):
      cv2.destroyAllWindows()
      break
cap.release()
```
